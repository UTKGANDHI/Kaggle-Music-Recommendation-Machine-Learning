```python
import numpy as np 
import pandas as pd 
from sklearn.model_selection import KFold

#reading test and train data as dataframes named df_test and df_train respectively

df_test = pd.read_csv('file:///E:/Namrata/study/cs/1stsem/MachineLearning/Project/datasets/test.csv',dtype={'msno' : 'category',
'source_system_tab' : 'category','source_screen_name' : 'category', 'source_type' : 'category','song_id' : 'category'})

df_train = pd.read_csv('file:///E:/Namrata/study/cs/1stsem/MachineLearning/Project/datasets/train.csv',dtype={'msno' : 'category',
'source_system_tab' : 'category',
                                                  'source_screen_name' : 'category',
                                                  'source_type' : 'category',
                                                  'target' : np.uint8,
                                                  'song_id' : 'category'})


```

```python
#using members datasets

df_members = pd.read_csv('file:///E:/Namrata/study/cs/1stsem/MachineLearning/Project/datasets/members.csv',dtype={'city' : 'category',
                                                      'bd' : 'category',
                                                      'gender' : 'category',
                                                      'registered_via' : 'category'})


```

```python
# creating a column of registration_init_date  using month, year and day

df_train.registration_init_time = pd.to_datetime(df_train.registration_init_time, format='%Y%m%d', errors='ignore')
df_train['registration_init_time_year'] = df_train['registration_init_time'].dt.year
df_train['registration_init_time_month'] = df_train['registration_init_time'].dt.month
df_train['registration_init_time_day'] = df_train['registration_init_time'].dt.day

df_members.drop('registration_init_time',axis=1,inplace=True)

```

```python

# creating expiration date using year, date, month

df_train.expiration_date = pd.to_datetime(df_train.expiration_date,  format='%Y%m%d', errors='ignore')
df_train['expiration_date_year'] = df_train['expiration_date'].dt.year
df_train['expiration_date_month'] = df_train['expiration_date'].dt.month
df_train['expiration_date_day'] = df_train['expiration_date'].dt.day

df_test = pd.merge(left = df_test,right = df_members,how='left',on='msno')
df_test.msno = df_test.msno.astype('category')
df_train = pd.merge(left = df_train,right = df_members,how='left',on='msno')
df_train.msno = df_train.msno.astype('category')

del df_members


```

```python
df_songs = pd.read_csv('file:///E:/Namrata/study/cs/1stsem/MachineLearning/Project/datasets/songs.csv',dtype={'genre_ids': 'category',
                                                  'language' : 'category',
                                                  'artist_name' : 'category',
                                                  'composer' : 'category',
                                                  'lyricist' : 'category',
                                                  'song_id' : 'category'})

df_test = pd.merge(left = df_test,right = df_songs,how = 'left',on='song_id')
df_test.song_length.fillna(200000,inplace=True)
df_test.song_length = df_test.song_length.astype(np.uint32)
df_test.song_id = df_test.song_id.astype('category')

df_train = pd.merge(left = df_train,right = df_songs,how = 'left',on='song_id')
df_train.song_length.fillna(200000,inplace=True)
df_train.song_length = df_train.song_length.astype(np.uint32)
df_train.song_id = df_train.song_id.astype('category')

del df_songs

```

```python
#implementing PCA

import numpy as np
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(df_train[df_train.columns[df_train.columns != 'target']])
```

```python
import lightgbm as lgbm
#K fold cross validation

k = KFold(n_splits=3)
X_array=np.zeros(shape=(len(df_train)))
predictions = np.zeros(shape=[len(df_test)])
X = df_train.drop(['target'], axis=1)
y = df_train['target'].values

for train_indices,val in k.split(df_train) : 
    td = lgbm.Dataset(df_train.drop(['target'],axis=1).loc[train_indices,:],label=df_train.loc[train_indices,'target'])
    vd = lgbm.Dataset(df_train.drop(['target'],axis=1).loc[val,:],label=df_train.loc[val,'target'])
    #X_array=X_array.append(df_train.loc[val,'target'])
    
    params = {
        'metric': 'binary_logloss',
        'boosting': 'gbdt',
        'learning_rate': 0.2 ,
        'verbose': 0,
        'num_leaves': 2**8,
        'bagging_fraction': 0.95,
        'bagging_freq': 1,
        'bagging_seed': 1,
        'feature_fraction': 0.9,
        'feature_fraction_seed': 1,
        'max_bin': 128,
        'max_depth': 10,
        'num_rounds': 200,
        'application':'binary'
        } 
    
    model = lgb.train(params, td, 100, valid_sets=[vd])
    predictions+=model.predict(df_test.drop(['id'],axis=1))
    
predictions = predictions/3
#taking the output in csv file known as submission.csv
submission = pd.read_csv('file:///E:/Namrata/study/cs/1stsem/MachineLearning/Project/datasets/sample_submission.csv')
submission.target=predictions
submission.to_csv('submission.csv',index=False)
```
